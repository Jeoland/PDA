```python
sentiment_map = {"0":"negative","1":"neutral","2":"positive"}

def read_tsv(domain,type):
    with open(f"{domain}/{type}.tsv", 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip().split("\t")
            sent = line[0]  #句子
            sent_split = sent.split(" ")
            qual = []
            cos_triples = []
            for q in line[1:]:
                aspect,category,sentiment,opinion = q.split(" ")
                asp_start,asp_end = aspect.split(",")
                opn_start,opn_end = opinion.split(",")
                senti = sentiment_map[sentiment]

                if asp_start == "-1" and asp_end == "-1":
                    continue
                if opn_start == "-1" and opn_end == "-1":
                    continue
                else:
                    asp = " ".join(sent_split[int(asp_start):int(asp_end)])
                    opn = " ".join(sent_split[int(opn_start):int(opn_end)])

                    qual.append((asp,category,senti,opn))  ##四元组
                    cos_triples.append((category,opn,senti))  ##三元组

            ## 如果不为空，则保存
            if cos_triples:
                with open(f"{domain}/{type}.txt", "a", encoding="utf-8") as out_f:
                    triples_str = str(cos_triples)
                    out_f.write(f"{sent}####{triples_str}\n")

            # print(sent,qual)
    return


domain = ['laptop',"rest"]
type = ['train','dev','test']
for d in domain:
    for t in type:
        read_tsv(d,t)
```
