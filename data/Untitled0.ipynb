```python
def upper_lower(dataset,data_type):
  data = open(f'./{dataset}/{data_type}','r').readlines()
  for line in data:
    sents,triplets = line.strip().split("####")
    new_t = []
    for t in eval(triplets):
      cate = " ".join([c.lower() for c in t[0].split("#")])
      new_t.append((cate,t[1],t[2]))
    s = f"{sents}####{new_t}"
    # print(s)

    ## 保存txt文件
    with open(f"{dataset}/new_{data_type}",'a')as file:
      file.write(s + '\n')
  print("saved finished")
  return

dataset = 'laptop'
data_type = 'dev.txt'
upper_lower(dataset,data_type)
```
